```python
import pandas as pd
from scipy.stats import pearsonr

df = pd.read_csv("data.csv", sep=";")

df.head()
```

```python
# show values of the target column
print("Target values:", df['Target'].unique())

# check datatypes
print("\nData types:")
print(df.dtypes)

# check for missing values in each column
print("\nMissing values:")
print(df.isna().sum())
```

```python
def encode_target(x):
    if x == "Graduate":
        return 1
    elif x == "Dropout":
        return 0
    else:
        return x

# removing enrolled because we don't know which students will drop out or graduate. 
# I feel like it will obfuscate any correlations
df = df[df['Target'].isin(['Dropout', 'Graduate'])]

# encoding the target column
df["Target"] = df["Target"].apply(encode_target)

print("Target values:", df['Target'].unique())

correlations = []

for column in df.columns:
    if column != "Target":
                # Check if p-value is below threshold (e.g., 0.05)
        corr, p_value = pearsonr(df['Target'], df[column])
        if (p_value < 0.05):
            correlations.append((column, corr, p_value))

correlations.sort(key=lambda x: x[1], reverse=True)

print("Number of statistically significant correlations:", len(correlations))
for corr in correlations:
    print(corr[0] + ": correlation ->", str(round(corr[1], 2)) + ", p-value ->", round(corr[2], 4))
```
